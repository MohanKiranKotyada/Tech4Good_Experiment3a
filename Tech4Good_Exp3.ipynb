```python
#Cost

import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# Load Excel file
file_path = 'Exp3_Cost Constraints.xlsx'
df = pd.read_excel(file_path)

# Extract sentences from the "ideas" column
sentences = df['ideas'].tolist()

# Load the sentence transformer model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Perform word embedding
embeddings = model.encode(sentences)

# Perform PCA to reduce dimensionality to 2D for plotting
pca = PCA(n_components=2)
reduced_embeddings = pca.fit_transform(embeddings)

# Plot the embedded vectors with numbers
plt.figure(figsize=(10, 8))
for i, (x, y) in enumerate(reduced_embeddings):
    plt.scatter(x, y)
    plt.text(x + 0.01, y + 0.01, str(i + 1), fontsize=9)
plt.title('PCA Plot of Embedded Vectors')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.grid(True)
plt.show()

# Compute the minimum distance between each point and previously generated points
min_distances = []
for i in range(1, len(reduced_embeddings)):
    current_point = reduced_embeddings[i]
    prev_points = reduced_embeddings[:i]
    distances = np.linalg.norm(prev_points - current_point, axis=1)
    min_distance = np.min(distances)
    min_distances.append(min_distance)

# Calculate moving average with a window size of 4
window_size = 4
moving_average = np.convolve(min_distances, np.ones(window_size)/window_size, mode='valid')

# Plot the minimum distances and moving average
plt.figure(figsize=(10, 8))
plt.plot(range(1, len(reduced_embeddings)), min_distances, marker='o', label='Minimum Distance')
plt.axvline(x=20, color='r', linestyle='--', label='Vertical Line After Point 20')
plt.plot(range(window_size, len(reduced_embeddings)), moving_average, color='r', label='Moving Average (window size = 4)')
plt.title('Minimum Distance to Previously Generated Point with Moving Average: For Cost Constraints')
plt.xlabel('Point Index')
plt.ylabel('Minimum Distance')
plt.legend()
plt.grid(True)
plt.show()
```
